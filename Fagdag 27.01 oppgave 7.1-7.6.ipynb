# Oppgave 7.1

```python
import pylab

def derivert(f, x, dx):
    fder = (f(x + dx) - f(x))/dx
    return fder

def f(x):
    return 2*x+1
def g(x):
    return x**2-4*x+5
def k(x):
    return pylab.e**x

xlist = []
flist = []
glist = []
klist = []
dx = 10**(-8)

for i in range(2, 5):
    fderivert = derivert(f, i, dx) 
    gderivert = derivert(g, i, dx) 
    kderivert = derivert(k, i, dx)
    flist.append(fderivert)
    glist.append(gderivert)
    klist.append(kderivert)
    xlist.append(i)
     
pylab.plot(xlist, flist)
pylab.plot(xlist, glist)
pylab.plot(xlist, klist)

pylab.show()
    
```

```python
data = pylab.loadtxt('posisjon.txt', skiprows = 1, delimiter = "        ")

tid = data[:,0]
pos = data[:,1]

aks = []
fart = []

for i in range(len(tid)-1):
    dy = pos[i+1]-pos[i]
    dx = tid[i+1]-tid[i]
    der = (dy)/(dx)
    fart.append(der)
    
for i in range(len(tid)-2):
    dy = fart[i+1]-fart[i]
    dx = tid[i+1]-tid[i]
    der = (dy)/(dx)
    aks.append(der)
    
pylab.plot(tid[:-1], fart)
pylab.plot(tid[:-2], aks)
pylab.show()

```

man tar utgangspunkt i en nummerisk verdi av x og man setter da delta x i liten, istedenfor tilnærmet null. I analytisk derivasjon prøver man å finne en ny funksjon, mens i nummerisk derivasjon prøver man å finne en verdi. Nummerisk integrasjon er å finne en tilnærmet verdi under grafen. Når delta x er liten vil gi en mer nøyaktig verdi av den deriverte enn dersom delta x er større. Dette er fordi avstanden mellom de to punktene man måler mindre, og derfor mere nøyaktig. Det samme gjelder ved integrasjon, bare i antall rektangler. Jo flere rektangler, jo mindre avstand vil det være fra rektangelene og grafen, og arealet vil være mer nøyaktig.
