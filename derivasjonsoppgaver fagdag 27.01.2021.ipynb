# Oppgave 7.1

```python
import pylab

def derivert(f, x, dx):
    fder = (f(x + dx) - f(x))/dx
    return fder

def f(x):
    return 2*x + 1

def g(x):
    return x**2 - 4*x + 5

def h(x):
    return pylab.e**x

print(h(2))


xlist = []
flist = []
glist = []
hlist = []
dx = 1E-8

for i in range(2, 5):
    fderivert = derivert(f, i, dx)
    gderivert = derivert(g, i, dx)
    hderivert = derivert(h, i, dx)
    flist.append(fderivert)
    glist.append(gderivert)
    hlist.append(hderivert)
    xlist.append(i)
    
pylab.plot(xlist, flist)
pylab.plot(xlist, glist)
pylab.plot(xlist, hlist)    

pylab.show()
```
